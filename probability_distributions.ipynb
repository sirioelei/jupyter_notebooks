## Gęstość rozkładu normalnego

Gęstość prawdopodobieństwa (PDF) rozkładu normalnego, czasami nazywana krzywą Gaussa lub dzwonem Gaussa, jest dana wzorem:

\[
f(x|\mu, \sigma^2) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
\]

gdzie:
- \( \mu \) to średnia (wartość oczekiwana) rozkładu normalnego,
- \( \sigma^2 \) to wariancja rozkładu normalnego,
- \( x \) to wartość dla której obliczamy gęstość prawdopodobieństwa.

Rozkład normalny jest symetryczny wokół średniej \( \mu \) i charakteryzuje się swoją charakterystyczną dzwonowatą krzywą. Im większa wariancja \( \sigma^2 \), tym bardziej spłaszczona jest krzywa.

W przypadku kodu, który generuje histogram rozkładu normalnego N(0,1), gdzie \( \mu = 0 \) i \( \sigma^2 = 1 \), wykorzystujemy bibliotekę `scipy.stats`, aby obliczyć gęstość prawdopodobieństwa na podstawie tego wzoru.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generowanie danych z rozkładu normalnego N(0,1)
data = np.random.normal(0, 1, 1000)

# Tworzenie histogramu
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Dodawanie linii dla gęstości prawdopodobieństwa
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)
title = "N(0,1)"
plt.title(title)
plt.show()
```
