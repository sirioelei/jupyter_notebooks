# Rozkład Poissona (rozkład dyskretny)

## Wzór rozkładu Poissona

Funkcja prawdopodobieństwa $ P(X = k) $ dla rozkładu Poissona jest dana wzorem:

$$
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
$$

gdzie:
- $ k $ to liczba zdarzeń,
- $ \lambda $ to średnia liczba zdarzeń występujących w danym przedziale czasu lub przestrzeni.

## Właściwości rozkładu Poissona

1. Średnia i wariancja rozkładu Poissona wynoszą $ \lambda $.
2. Rozkład Poissona jest dyskretny, co oznacza, że możliwe są tylko całkowite wartości liczbowe $ k $.
3. Rozkład Poissona jest skośny w prawo, szczególnie dla małych wartości $ \lambda $, a dla dużych wartości $ \lambda $ przypomina rozkład normalny.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry rozkładu Poissona
lambd = 5  # Średnia liczba zdarzeń
k = 500  # Liczba próbek

# Generowanie danych z rozkładu Poissona
data = np.random.poisson(lambd, k)

# Tworzenie histogramu
plt.hist(data, bins=150, density=False, alpha=0.7, color='g')

# Dodanie etykiet i tytułu
plt.xlabel('Wartość')
plt.ylabel('Częstość występowania')
plt.title('Histogram rozkładu Poissona dla lambda={} i k={}'.format(lambd, k))

# Wyświetlenie histogramu
plt.show()
```

# Rozkład normalny (rozkład ciągły)
## Gęstość rozkładu normalnego

Gęstość prawdopodobieństwa (PDF) rozkładu normalnego, czasami nazywana krzywą Gaussa lub dzwonem Gaussa, jest dana wzorem:
$$
    f(x|\mu, \sigma^2) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

gdzie:
- $ \mu $ - średnia (wartość oczekiwana) rozkładu normalnego,
- $ \sigma^2 $ - wariancja rozkładu normalnego,
- $ x $ - wartość dla której obliczamy gęstość prawdopodobieństwa.

Rozkład normalny jest symetryczny wokół średniej $ \mu $ i charakteryzuje się swoją charakterystyczną dzwonowatą krzywą. Im większa wariancja $ \sigma^2 $, tym bardziej spłaszczona jest krzywa.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generowanie danych z rozkładu normalnego N(0,1)
data = np.random.normal(0, 1, 1000)

# Tworzenie histogramu
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Dodawanie linii dla gęstości prawdopodobieństwa
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)
title = "N(0,1)"
plt.title(title)
plt.show()
```

# Rozkład Wykładniczy

## Wzór rozkładu wykładniczego - funkcja gęstości

Funkcja gęstości prawdopodobieństwa (PDF) rozkładu wykładniczego jest dana wzorem:

$$
    f(x|\lambda) = \lambda e^{-\lambda x}
$$

gdzie:
- $ \lambda $ to parametr rozkładu, nazywany również częstością występowania zdarzeń.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parametr rozkładu wykładniczego
lambd = 0.5  # Częstość występowania zdarzeń

# Zakres wartości
x = np.linspace(0, 10, 100)

# Funkcja gęstości prawdopodobieństwa
pdf = lambd * np.exp(-lambd * x)

# Tworzenie wykresu
plt.plot(x, pdf, label='PDF')
plt.xlabel('x')
plt.ylabel('PDF(x)')
plt.title('Rozkład Wykładniczy (lambda={})'.format(lambd))
plt.legend()
plt.grid(True)
plt.show()
```

# Wielowymiarowy Rozkład Normalny

Wielowymiarowy rozkład normalny, znany również jako wielowymiarowy rozkład Gaussa, jest generalizacją jednowymiarowego rozkładu normalnego na większą liczbę wymiarów. Jest to często używany model w statystyce i uczeniu maszynowym, szczególnie w kontekście analizy danych wielowymiarowych.

## Funkcja Gęstości Prawdopodobieństwa

Funkcja gęstości prawdopodobieństwa (PDF) wielowymiarowego rozkładu normalnego jest dana wzorem:

$$
    f(x|\mu, \Sigma) = \frac{1}{(2\pi)^{n/2}|\Sigma|^{1/2}} e^{-\frac{1}{2}(x-\mu)^T \Sigma^{-1} (x-\mu)}
$$

gdzie:
- $ x $ to wektor zmiennych losowych wielowymiarowego rozkładu normalnego,
- $ \mu $ to wektor średnich wartości (wartości oczekiwane) dla każdej zmiennej,
- $ \Sigma $ to macierz kowariancji dla zmiennych losowych.

### Właściwości wielowymiarowego rozkładu normalnego

1. Wielowymiarowy rozkład normalny jest w pełni określony przez wektor średnich $ \mu $ i macierz kowariancji $ \Sigma $.
2. Dla wielowymiarowego rozkładu normalnego, wektor średnich $ \mu $ określa środek rozkładu, a macierz kowariancji $ \Sigma $ określa jego kształt.
3. Rozkład normalny wielowymiarowy jest symetryczny wokół wektora średnich $ \mu $, co oznacza, że największe wartości gęstości prawdopodobieństwa występują wokół $ \mu $.
4. Zmienna losowa $ x $ ma wielowymiarowy rozkład normalny, jeśli każda kombinacja liniowa jej komponentów ma jednowymiarowy rozkład normalny.
